                 a. Create a new folder

                 b. Put the following files in the folder


                                    Code.txt

                                    Log.txt

                                    Output.txt


                 c. Stage the Code.txt and Output.txt files

                 d. Commit them

                 e. And finally push them to GitHub

*Ans.*

2. Initialize Git -	git init	- Initializes the repository.

3. Create Files	- touch Code.txt	- all 3 file create same command

4. Stage Files	- git add Code.txt Output.txt -	Stages only Code.txt and Output.txt.

7. Set Branch -	git branch -M main	- Renames the local branch to main.

8. Push to GitHub -	git push -u origin main	- Pushes the commit to GitHub.

Q.2 - Tasks to Be Performed:


1. Create a Git working directory with feature1.txt and feature2.txt in the master branch

2. Create 3 branches develop, feature1 and feature2

3. In develop branch create develop.txt, do not stage or commit it

4. Stash this file and check out to feature1 branch

5. Create new.txt file in feature1 branch, stage and commit this file

6. Checkout to develop, unstash this file and commit

7. Please submit all the Git commands used to do the above step>

*Ans.*

1. Initial Setup and Master Branch Commit

# Create the directory and initialize Git
mkdir devops-task-2
cd devops-task-2
git init

# Create and commit initial files to master/main
echo "Master feature 1 content" > feature1.txt
echo "Master feature 2 content" > feature2.txt
git add .
git commit -m "Initial commit with master files"

2. Create Branches and Switch to develop

# Create the branches
git branch develop
git branch feature1
git branch feature2

# Switch to the development branch
git checkout develop

3. Stash Uncommitted Work and Switch to feature1

# Create the file (Unstaged)
echo "Develop file content" > develop.txt

# Stash the untracked file and all current working directory changes
git stash push -u

# Switch to the feature branch
git checkout feature1

4. Commit in feature1 Branch

# Create, stage, and commit the new file in feature1
echo "New content for feature 1" > new.txt
git add new.txt
git commit -m "Added new.txt in feature1 branch"

5. Switch back to develop, Unstash, and Commit

# Switch back to the development branch
git checkout develop

# Retrieve the stashed file (develop.txt) and remove the stash entry
git stash pop

# Stage and commit the retrieved file
git add develop.txt
git commit -m "Added develop.txt after unstashing"

Q.5 - Tasks to Be Performed:

1. Create a Git Flow workflow architecture on Git

2. Create all the required branches

3. starting from e feature branch, push the branch to the master, following the architecture

4. Push an urgent.txt on master using hotfix

*Ans.*

# A. Initial Setup and Develop Branch Creation (Tasks 1 & 2)
mkdir devops-task-5
cd devops-task-5
git init

# Initial commit to master
echo "Initial Project Setup" > project_init.txt
git add .
git commit -m "Initial commit to master"

# Create the primary permanent branch
git branch develop
git checkout develop

# B. Feature Branch Workflow and Push to Master (Task 3)

# Start a feature branch from develop
git checkout -b feature/new-login
echo "Login feature code" > feature_login.txt
git add .
git commit -m "Added login feature"

# Merge feature into develop
git checkout develop
git merge feature/new-login
git branch -d feature/new-login

# Create Release branch from develop (to transition the feature to master)
git checkout -b release/1.0.0

# Merge Release into master (This is the push to master)
git checkout master
git merge release/1.0.0

# Delete local release branch (optional)
git branch -d release/1.0.0

# C. Hotfix Workflow (Task 4)

# 4. Push an urgent.txt on master using hotfix
# Start hotfix branch from master
git checkout -b hotfix/urgent-fix master
echo "URGENT FIX: Production issue resolved" > urgent.txt
git add .
git commit -m "HOTFIX: Added urgent.txt to production"

# Merge hotfix into master (Fix applied to production)
git checkout master
git merge hotfix/urgent-fix

# Merge hotfix into develop (Keeps development up-to-date)
git checkout develop
git merge hotfix/urgent-fix

# Delete local hotfix branch (optional)
git branch -d hotfix/urgent-fix
